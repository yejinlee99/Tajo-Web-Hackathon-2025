```python
# llama3
import requests

url = "http://localhost:11434/api/generate"
prompt_text = result_stt_one["text"]  # Whisper ê²°ê³¼ì—ì„œ í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ
payload = {
    "model": "llama3",           # ë„ˆê°€ ì‹¤í–‰ ì¤‘ì¸ ëª¨ë¸ ì´ë¦„
    "prompt": prompt_text,
    "stream": False,
    "device" : "cuda" # gpu ê°€ë™ ì˜µì…˜
}

response = requests.post(url, json=payload)
result_llm = response.json()

# ì „ì²´ ì‘ë‹µ í™•ì¸
import json
print("ğŸ” ì‘ë‹µ ë‚´ìš© ì „ì²´:")
print(result_llm)
# print(json.dumps(result, indent=2))
```
